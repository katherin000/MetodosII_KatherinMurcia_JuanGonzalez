```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x*(y**2) - 2*y/x - 1/(x**3)


def sol(x):
    return 2/(x**2 * (x**(2) + 2)) - 1/(x**2 + 2)


def RK4(x, y, h):

    for i in range(1, len(y)):

        k1 = f(x[i-1], y[i-1])
        k2 = f(x[i-1] + 0.5*h, y[i-1] + 0.5*k1*h)
        k3 = f(x[i-1] + 0.5*h, y[i-1] + 0.5*k2*h)
        k4 = f(x[i-1] + h, y[i-1] + h*k3)

        y[i] = y[i-1] + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
```

```python
N = 50
t = np.linspace(np.sqrt(2), 5 + np.sqrt(2), N)
h = t[1] - t[0]

y = np.zeros_like(t)
y[0] = 0.

RK4(t, y, h)
```

```python
plt.figure(figsize = (10, 7))
plt.plot(t, sol(t), color='y', label='Solución')
plt.plot(t, y, linestyle=':', color='b', label='Runge-Kutta 4')
plt.legend()
```
