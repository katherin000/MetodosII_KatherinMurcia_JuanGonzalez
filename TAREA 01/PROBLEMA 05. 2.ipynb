```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Euler(a, u0, dt, k):
    ks = np.linspace(0, k-1, k)
    sol = np.zeros(k)
    for i in ks:
        sol[int(i)] = u0*(1 + a*dt)**i
    return sol
```

```python
a = -5
k = 100
u0 = 3


t = np.linspace(0,10,k)



func = u0 * np.exp(a * t)





fig1 = plt.figure(figsize=(12,7))
ax1 = fig1.add_subplot(131)
ax2 = fig1.add_subplot(132)
ax3 = fig1.add_subplot(133)
#ax4 = fig1.add_subplot(224)


ax1.plot(func, t, label= 'Exacta')
ax1.plot(Euler(a, u0, (-0.9/a), k), t, label='Numérica')
ax1.set_title('α = -5, Δ t < -1/α')
ax1.legend()



ax3.plot(func, t, label= 'Exacta')
ax3.plot(Euler(a, u0, (-1.1/a), k), t, label='Numérica')
ax3.set_title('α = -5, Δ t > -1/α')
ax3.legend()

ax2.plot(func, t, label= 'Exacta')
ax2.plot(Euler(a, u0, (-1/a), k), t, label='Numérica')
ax2.set_title('α = -5, Δ t = -1/α')
ax2.legend()





```

```python
a = -1

real = u0 * np.exp(a * t)

dt = np.array([ 1.1, 1.5, 1.9])

euls1 = []
for i in dt:
    eul1 = Euler(a, u0, i, k)
    euls1.append(eul1)


fig2 = plt.figure(figsize=(12,7))
ax5 = fig2.add_subplot(131)
ax6 = fig2.add_subplot(132)
ax7 = fig2.add_subplot(133)

ax5.plot(real, t, label='Exacta')
ax5.plot(euls1[0], t, label='Numérica')
ax5.set_title('α = -1, Δ t = 1.1')
ax5.legend()

ax6.plot(real, t, label='Exacta')
ax6.plot(euls1[1], t, label='Numérica')
ax6.set_title('α = -1, Δ t = 1.5')
ax6.legend()

ax7.plot(real, t, label='Exacta')
ax7.plot(euls1[2], t, label='Numérica')
ax7.set_title('α = -1, Δ t = 1.9')
ax7.legend()

```
